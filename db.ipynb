```python
def clean_phone_number(x):
    prefixes = ['7','8','9']
    x = str(x)
    x = x.lower()
    x = x.replace(" ", "").replace("or", ";").replace("o", "0").replace("-", "").replace("Â±", "+").replace(",", "")
    x = re.split(';|,|/', x)[0]
    if x.startswith(tuple(prefixes)):
        x = "0" + x
    if x.startswith("0"):
        if len(x)>11:
            x=x[:11]
    if len(x)<10:
        x=""
    if len(x)>14:
        x=x[:14]
    if "e" in x:
        x=""
    return x

df['Phone number'] = df['Phone number'].apply(clean_phone_number)
```

```python
import datetime
ranges = [(1,5),(6,10),(11,15),(16,20),(21,25),(26,30),(31,35),(36,40),(41,45),(46,50),(51,55),(56,60),(61,65)]
def check_int(s):
    # if s[0] in ('-', '+'):
    #     return s[1:].isdecimal()
    return s.isdecimal()

def age_from_dob(x):
    if x:
        x=str(x)
        try:
            x=datetime.datetime.strptime(x,'%m/%d/%Y')
        except ValueError as e:
            x=datetime.datetime.strptime(x,'%Y-%m-%d %H:%M:%S')
        except Exception as e:
            raise e
        if x.year < 1900:
            x = x.replace(year=x.year+1900)
        return x

def age_range(row):
    if type(row['Date of Birth']) == pd._libs.tslibs.nattype.NaTType:
        pass
    elif row['Age Range'] and len(str(row['Age Range']).split(' ')) == 1:
        y = str(row['Age Range']).split(' ')[0]
        if check_int(str(y)):
            y= int(y)
            for r in ranges:
                if r[0] <= y and r[1] >= y:
                    return f'{r[0]} - {r[1]}'
    elif row['Date of Birth']:
        years = int((datetime.datetime.now() - row['Date of Birth']).days/365.25)
        for r in ranges:
            if r[0] <= years and r[1] >= years:
                return f'{r[0]} - {r[1]}'
    return row['Age Range']
```

```python
df['Date of Birth'] = df['Date of Birth'].apply(age_from_dob)
```

```python
df['Date of Birth'].fillna("", inplace=True)
df['Age Range'] = df.apply(lambda row: age_range(row), axis=1)
```

```python
def last_name(row):
    if row['Surname']:
        return row['Surname']
    name = row['First Name'].strip().split(" ")
    return name[-1]

def first_name(row):
    fn =  row['First Name'].strip().split(" ")
    if len(fn) >= 2:
        fn = fn[:-1]
    return " ".join(fn)
```

```python
df['Surname'] = df.apply(lambda row: last_name(row), axis=1)
df['First Name'] = df.apply(lambda row: first_name(row), axis=1)
```

```python
df['Gender'] = df['Gender'].apply(lambda x: x.upper())
```

```python
def loe(value):
    if value == "Bachelor's Degree":
        return 'BSC'
    elif value == "Elementary School":
        return 'ELEMENTRY_SCHOOL'
    elif value == "Secondary School":
        return 'SECONDARY_SCHOOL'
    elif value == "College of Education":
        return 'COLLEGE_OF_EDUCATION'
    elif value == "ND/HND":
        return 'ND_HND'
    elif value == "Master's Degree":
        return 'MSC'
    else:
        return ''

def ca(value):
    if value == 'Urban':
        return 'URBAN'
    elif value == 'Rural':
        return 'RURAL'
    elif value == 'Peri-Urban':
        return 'PERI_URBANS'
```
