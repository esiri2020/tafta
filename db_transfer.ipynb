```python
# Get the list of tables to copy from the source database
with source_conn.cursor() as source_cur:
    source_cur.execute("SELECT table_name FROM information_schema.tables WHERE table_schema = 'public' AND table_type = 'BASE TABLE'")
    tables_to_copy = [row[0] for row in source_cur.fetchall()]
```

```python
ordered_table_names = []
while len(tables_to_copy) > 0:
    for table_name in tables_to_copy:
        if table_name == 'spatial_ref_sys':
            tables_to_copy.remove(table_name)
        else:
            with source_conn.cursor() as source_cur:
                source_cur.execute("SELECT conname, conrelid, confrelid FROM pg_constraint WHERE confrelid = (SELECT oid FROM pg_class WHERE relname = %s)", (table_name,))
                fk_constraints = source_cur.fetchall()
                # Check if all foreign key dependencies have already been added to the ordered list
                all_deps_added = all([fk_constraint[1] in [dest_conn.cursor().execute("SELECT oid FROM pg_class WHERE relname = %s", (name,)).fetchone()[0] for name in ordered_table_names] for fk_constraint in fk_constraints])
                if all_deps_added:
                    ordered_table_names.append(table_name)
                    tables_to_copy.remove(table_name)
                else:
                    break
```

```python
# Iterate over the ordered list of tables and copy the data
for table_name in ordered_table_names:
    print(table_name)
    with source_conn.cursor() as source_cur, dest_conn.cursor() as dest_cur:
        # Get the column names of the table
        source_cur.execute("SELECT column_name FROM information_schema.columns WHERE table_name = %s", (table_name,))
        column_names = [f'"{row[0]}"' for row in source_cur.fetchall()]
        print(column_names)
        # Build the select statement
        select_statement = f'SELECT * FROM "{table_name}"'
        
        # Execute the select statement
        source_cur.execute(select_statement)
        rows = source_cur.fetchall()
        
        # Build the insert statement
#         insert_statement = f'INSERT INTO public."{table_name}" ({",".join(column_names)}) SELECT {",".join(column_names)} FROM "{table_name}"'
        
        # Execute the insert statement
        for row in rows:
            dest_cur.execute(f'INSERT INTO "{table_name}" ({",".join(column_names)}) VALUES ({",".join(["%s" for s in column_names])})', row)
        
    # Commit the changes
    dest_conn.commit()
```
